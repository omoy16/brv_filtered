```python
import pandas as pd
from itertools import combinations
from src.analysis.univariate import analyze_pair

datasets_path = r'C:\Users\omoy1\Desktop\KABUCALIB\datasets\brv_pending_datasets.csv'
features_df = pd.read_csv(datasets_path)
features_df.head()

df = features_df.copy()

# feature_cols の定義が必要なため、ここでは仮に数値列から抽出（必要に応じて調整してください）
feature_cols = features_df.select_dtypes(include=["number"]).columns.tolist()
exclude_cols = ["Target_Profit", "Profit_Pct"] # 必要に応じて追加
feature_cols = [col for col in feature_cols if col not in exclude_cols]

for f1, f2 in combinations(feature_cols, 2):
    res = analyze_pair(df, f1, f2)
    print("=" * 80)
    print(f"{f1} × {f2}")
    print("\n【WinRate】")
    print(res.winrate_df)
    print("\n【Count】")
    print(res.count_df)
```

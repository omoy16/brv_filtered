# Univariate Analysis Demo
このNotebookは、`src.analysis.univariate` ライブラリを使用して特徴量の単変量分析を行い、その結果を可視化・検証するためのデモ用Notebookです。

```python
import pandas as pd
import sweetviz as sv
from src.analysis.univariate import analyze_features, export_univariate_results

# データロード
datasets_path = r'C:\Users\omoy1\Desktop\KABUCALIB\datasets\brv_pending_datasets.csv'
features_df = pd.read_csv(datasets_path)
features_df.head()
```

```python
# 分析対象特徴量の選定
feature_cols = features_df.select_dtypes(include=["number"]).columns.tolist()

exclude_cols = [
    "Target_Profit",
    "Profit_Pct",
    "Buy_Price",
    "Filter_Relative_Pos",
    "Detect_Price",
    "Last_Price",
]
feature_cols = [col for col in features_df.columns if col in feature_cols and col not in exclude_cols]

# 一括分析とCSVエクスポート
results = analyze_features(features_df, feature_cols)
saved_files = export_univariate_results(results)
print(f"Saved {len(saved_files)} files to outputs/univariate/")

# 結果の表示
for res in results:
    print(f"\n=== {res.feature_name} ===")
    display(res.summary_df)
```

## 外部ライブラリによる可視化 (Sweetviz)

```python
report = sv.analyze(features_df[feature_cols + ["Target_Profit"]])
report.show_html()
```
